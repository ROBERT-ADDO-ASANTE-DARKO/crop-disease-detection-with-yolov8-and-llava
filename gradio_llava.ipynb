```python
import gradio as gr
import numpy as np
from PIL import Image
from transformers import BitsAndBytesConfig, pipeline
from ultralytics import YOLO
#from transformers import LlavaForConditionalGeneration
import torch
```

```python
quant_config = BitsAndBytesConfig(
    load_in_4bit = True,
    bnb_4bit_compute_dtype = torch.float16
)
```

```python
model_id = "llava-hf/llava-1.5-7b-hf"
#pipe = pipeline("image-to-text", model=model_id, device=0)
#model = LlavaForConditionalGeneration.from_pretrained(model_id)

pipe = pipeline(
    "image-to-text",
    model=model_id,
    model_kwargs={"quantization_config": quant_config}
)
```

```python
# Load the YOLOv8 model (example, replace with your YOLO model loading logic)
yolo_model = YOLO("/content/drive/MyDrive/best.pt")  # Replace with your YOLO model loading code
#yolo_model = YOLO("yolov8m.pt")

def clean_text(text):
  # Remove special tokens or unwanted symbols using regular expressions
  cleaned_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
  return cleaned_text

import re
from PIL import Image

import re
from PIL import Image

def process_inputs(img, conf_threshold, iou_threshold, text):
    """Process image and text inputs."""

    if yolo_model is not None:
        results = yolo_model.predict(
            source=img,
            conf=conf_threshold,
            iou=iou_threshold,
            show_labels=True,
            show_conf=True,
            imgsz=640,
        )
        for r in results:
            im_array = r.plot()
            im = Image.fromarray(im_array)
    else:
        im = Image.fromarray(img)

    max_new_tokens = 500
    prompt_instructions = """
    Describe the image using as much detail as possible.
    You are a helpful AI assistant who is able to answer questions about the image.
    What is the image all about?
    Now generate the helpful answer.
    """

    prompt = "User: <image>\n" + prompt_instructions + "\nAssistant:"

    llava_response = pipe(im, prompt=prompt, generate_kwargs={"max_new_tokens": max_new_tokens})

    if llava_response and llava_response[0]["generated_text"]:
        match = re.search(r"Assistant:(.*)", llava_response[0]["generated_text"], re.DOTALL)
        if match:
            response = match.group(1).strip()
        else:
            response = llava_response[0]["generated_text"].strip()
    else:
        response = "No response generated."

    return im, response

# Make sure to define yolo_model and pipe before using this function.

iface = gr.Interface(
    fn=process_inputs,
    inputs=[
        gr.Image(type="numpy", label="Upload Image"),
        gr.Slider(minimum=0, maximum=1, value=0.25, label="Confidence threshold"),
        gr.Slider(minimum=0, maximum=1, value=0.45, label="IoU threshold"),
        #gr.Textbox(label="Chat with LLava"),
    ],
    outputs=[
        gr.Image(type="pil", label="Result"),
        gr.Textbox(label="LLava Response"),
    ],
    title="Crop Disease Detection with YOLOv8 and LLava",
    description="Upload your image to detect the plant disease.",
    examples=[
        ["https://ultralytics.com/images/zidane.jpg", 0.25, 0.45, "What's the weather like today?"],
        ["https://ultralytics.com/images/bus.jpg", 0.25, 0.45, "Hello, how are you?"],
    ],
)

if __name__ == "__main__":
    iface.launch()
```
